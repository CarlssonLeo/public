```python
import numpy as np
def find_max_sum(numbers):
    numbers.sort()
    return sum(numbers[-2:])

def find_max_sum_np(numbers):
    num = np.array(numbers)
    np.sort(num)
    return np.sum(num[-2:])

```

```python
import heapq
def find_max_sum2(numbers): 
    return sum(heapq.nlargest(2, numbers))
```

```python
def find_max_sum3(numbers):
    largest = 0
    second_largest = 0
    for item in numbers:
        if item > largest:
            second_largest = largest
            largest = item
        elif largest > item > second_largest:
            second_largest = item
    return largest+second_largest+1

```

```python
import random
randomlist = []
for i in range(1,100000):
    n = random.randint(1,4000)
    randomlist.append(n)



```

```python
find_max_sum(randomlist)
```

```python
find_max_sum2(randomlist)
```

```python
find_max_sum3(randomlist)
```

```python
find_max_sum_np(randomlist)


```
